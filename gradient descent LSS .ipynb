```python
m = 100
n = 50
```

```python
import numpy as np
import random
np.random.seed(27)
```

```python
A = np.random.randn(m,n)
y = np.random.randn(m)
x = np.random.randn(n)
```

```python
def LSS(A,x,y):
    z = y - np.matmul(A,x)
    return np.square(np.linalg.norm(z))

def gradient(A,x):
    return -2 * np.matmul(A.T, y - np.matmul(A,x))
```

```python
obj_val = []
epsilon = 0.001
t = 1
alpha = 0.2
beta = 0.5
while np.linalg.norm(gradient(A,x)) > epsilon:
    while LSS(A,x - (t * gradient(A,x)),y) >= (LSS(A,x,y) - (alpha * t * np.matmul(gradient(A,x),gradient(A,x).T))) :
        t = beta * t
    t_k = t
    x_new = x - (t_k * gradient(A,x))
    x = x_new
    obj_val.append(LSS(A,x,y))
```

```python
obj_val
```

```python
a = list(range(1,len(obj_val)+1))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,8))
plt.plot(a,obj_val,color = 'red')
plt.xlabel("Number of Iterations")
plt.ylabel("Objective value")
plt.title("Objective value vs Iterate")
plt.show()
```

As we can see the gradient descent method with backtracking line search conversges around 150 iterations where the gradient value approximately remains the same throughout for this problem.
